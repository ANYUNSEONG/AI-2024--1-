## 학습목표
 1. dictionary, set에 대해 이해 및 실습


### **dictionary**
 + 키와 값을 갖는 데이터 구조
 + 키는 내부적으로 hash값으로 저장(저장했을때는 내부적으론 인덱스로 전환)
 + 순서를 따지지 않음. 즉, 인덱스가 없음
 + 동일한 키가 중복될 수 없음

+ **항목 추가 및 변경**
 - 기존에 키가 존재 하면, 새로운 값으로 업데이트
 - 존재하지 않으면, 새로운 키, 값 생성

+ **update()**
 * 두 딕셔너리를 병합함
 * 겹치는 키가 있다면 parameter로 전달되는 키 값이 overwrite된다.

* **key삭제**
 - del 키워드 사용
 - pop 함수 사용

* **clear()**
 + 딕셔너리의 모든 값을 초기화


* **in**
 + key값 존재 확인
 + O(1) 연산 - 딕셔너리의 크기와 관계없이 항상 연산의 속도가 일정하다는 의미
 + in 키워드는 dictionary에서 중요함. 기능상에 차이는 없지만 리스트와 차이점은 성능 상에 차이가 있음.

* **value access**
 + dict[key]로 접근, 키가 없는 경우 에러 발생
 + .get() 함수로 접근, 키가 없는 경우 프로그램이 죽지않고 None반환

#### **모든 keys, values 접근**
 + keys() - 키만 반환
 + values() - 값만 반환
 + items() - 키, 값의 튜플을 반환

### **set**
 + dictionary에서 key만 활용하는 데이터 구조로 이해
 + 수학에서의 집합과 동일한 개념
 + 쓸 일이 많지 않음

* **set()으로 집합으로 변환**

+ **set operations**
 - 수학 연산과 동일
 - 교집합, 합집합, 차집합 등 지원
